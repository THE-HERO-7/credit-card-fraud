```python
dt=dt.drop_duplicates(subset=dt.columns.difference(['Time', 'Amount']))
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(dt.corr(),fmt=".2f", cmap="coolwarm",annot=True)
```

```python
fraud = dt[dt['Class'] == 1]
non_fraud = dt[dt['Class'] == 0]
print(f"Fraud cases: {len(fraud)}")
print(f"Non-fraud cases: {len(non_fraud)}")
non_fraud_sampled = non_fraud.sample(n=200000, random_state=42)
balanced_data = pd.concat([fraud, non_fraud_sampled], axis=0)
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)
print(balanced_data['Class'].value_counts())
```

```python
x=balanced_data.drop('Class',axis=1)
y=balanced_data.Class
x.shape,y.shape
frd = balanced_data[balanced_data['Class'] == 1]
non_frd =  balanced_data[balanced_data['Class'] == 0]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,stratify=y,random_state=42)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=200,random_state=42,n_jobs=-1)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, roc_auc_score,confusion_matrix,accuracy_score,roc_curve
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(roc_auc_score(y_test, y_pred))
```

```python
fpr,tpr,_=roc_curve(y_test,y_pred)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.show()
```
